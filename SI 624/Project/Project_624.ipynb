```python
diabetic_medication = [
    'glipizide', 'glimepiride', 'glyburide', 'pioglitazone', 'rosiglitazone',
    'acarbose', 'miglitol', 'tolazamide', 'insulin', 'glyburide-metformin',
    'glipizide-metformin', 'glimepiride-pioglitazone', 'metformin-rosiglitazone',
    'metformin-pioglitazone', 'metformin', 'repaglinide', 'nateglinide',
    'chlorpropamide', 'troglitazone', 'tolbutamide', 'acetohexamide', 'diabetic'
]

# Convert the "Brnd_Name" and "Gnrc_Name" column to lowercase
medicare['Brnd_Name_Lower'] = medicare['Brnd_Name'].str.lower()
medicare['Gnrc_Name_Lower'] = medicare['Gnrc_Name'].str.lower()

# Create a regular expression pattern to match any of the medication names anywhere in the string
medication_pattern = '|'.join(fr'.*{medication}.*' for medication in diabetic_medication)

# Create a new DataFrame containing rows with medication names matching the pattern
diabetic_medications = medicare[
    medicare['Gnrc_Name_Lower'].str.contains(medication_pattern, case=False, na=False) | medicare['Brnd_Name_Lower'].str.contains(medication_pattern, case=False, na=False) ]

```

```python
medication_counts = diabetic_medications['Gnrc_Name_Lower'].value_counts()

medication_counts
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
medication_counts.plot(kind='bar')
plt.title('Distribution of Diabetic Medications')
plt.xlabel('Medication')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Define a custom formatting function
def format_number(x):
    if x < 1e3:
        return x
    elif x < 1e6:
        return x / 1e3
    elif x < 1e9:
        return x / 1e6
    else:
        return x / 1e9

# Apply the custom formatting function to the describe result
description = medicare[selected_columns].describe().applymap(format_number)

# Create a table-like visualization (heatmap) using a color map
plt.figure(figsize=(10, 6))

# Convert the formatted numbers to a numeric format
description_numeric = description.apply(pd.to_numeric)

sns.heatmap(description_numeric, annot=True, fmt=".2f", cmap="YlGnBu", cbar=False, linewidths=0.5)
plt.title('Summary Statistics')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the "Tot_Spndng" columns for analysis
spending_columns = ['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']

# Create a pairplot to visualize the relationships between these columns
sns.pairplot(medicare[spending_columns], diag_kind='kde')
plt.suptitle('Pairplot of Tot_Spndng Columns')
plt.show()

# Create box plots for each year's total spending
plt.figure(figsize=(12, 6))
sns.boxplot(data=medicare[spending_columns])
plt.title('Box Plot of Tot_Spndng Columns')
plt.xticks(rotation=45)
plt.show()

# Select the "Tot_Spndng" columns for analysis
spending_columns = ['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']

# Aggregate the data to get the mean spending for each year
spending_summary = medicare[spending_columns].mean()

# Create a line plot to visualize the trend of total spending over the years
years = range(2017, 2022)

plt.figure(figsize=(10, 6))
sns.lineplot(x=years, y=spending_summary)
plt.title('Total Spending Over the Years')
plt.xlabel('Year')
plt.ylabel('Mean Total Spending')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the data for diabetic medication
diabetic_medication_spending = medicare[medicare['Gnrc_Name_Lower'].str.contains(medication_pattern, case=False, na=False) | medicare['Brnd_Name_Lower'].str.contains(medication_pattern, case=False, na=False)]

# Select the columns for years (e.g., Tot_Spndng_2017, Tot_Spndng_2018, etc.)
year_columns = ['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']

# Create lists to store years and corresponding total spending
years = []
total_spending = []

# Calculate total spending for each year and append to the lists
for year_column in year_columns:
    year = year_column.split('_')[2]
    total = diabetic_medication_spending[year_column].sum() / 1e9
    years.append(year)
    total_spending.append(total)
    print(f'Total spending in {year}: ${total:.2f}B')

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=years, y=total_spending, palette='Set3')

# Customize the plot
plt.title('Diabetic Medication Spending Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Spending (Billions)')
plt.show()
```
